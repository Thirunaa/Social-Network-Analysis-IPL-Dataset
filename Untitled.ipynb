```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
ball_by_ball_df = pd.read_csv('Ball_By_Ball.csv')
```

```python
match_df = pd.read_csv('Match.csv')

team_df = pd.read_csv('Team.csv')
```

```python
player_df = pd.read_csv('Player.csv', encoding="unicode_escape")
player_match_df = pd.read_csv('Player_match.csv', encoding="unicode_escape")
```

```python
nodes_df = ball_by_ball_df[['MatcH_id', 'Over_id','Ball_id','Innings_No','Striker','Bowler']]
```

```python
nodes_df
```

```python
nodes_df.rename({'Striker': 'PLAYER_SK'}, axis=1, inplace=True)
```

```python
nodes_df
```

```python
nodes_df['Bowler'] = nodes_df['Bowler'].map(player_df.set_index('PLAYER_SK')['Player_Name'])
```

```python
nodes_df
```

```python
edges_df = ball_by_ball_df[['MatcH_id', 'Over_id','Ball_id','Innings_No','Striker','Bowler','Runs_Scored','Bowler_Wicket']]
```

```python
edges_df
```

```python
edges_df['Bowler'] = edges_df['Bowler'].map(player_df.set_index('PLAYER_SK')['Player_Name'])
```

```python
edges_df['Striker'] = edges_df['Striker'].map(player_df.set_index('PLAYER_SK')['Player_Name'])
```

```python
edges_df
```

```python
edges_df.groupby(['Striker', 'Bowler']).agg({'Runs_Scored':'sum','Bowler_Wicket':'sum'})
```

```python
edges_df.to_csv('Nodes_edges.csv', index=False)
```

```python
edges_runs_wickets_df = edges_df.groupby(['Striker', 'Bowler'])[['Runs_Scored', 'Bowler_Wicket']].sum().reset_index()
```

```python
edges_runs_wickets_df.to_csv('edges_runs_wickets_df.csv', index=False)
```
